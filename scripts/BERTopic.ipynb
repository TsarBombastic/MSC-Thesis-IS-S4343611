```python
# File name: BERTopic.ipynb
# This program allows for topic modeling of data
# Author: Maxim van der Maesen de Sombreff
# Date: 29-06-2025

from bertopic import BERTopic
import openai
from bertopic.representation import OpenAI
import os
import pandas as pd
from transformers.pipelines import pipeline

"""
This script allow for topic modeling using BERTopic, fine-tuned on DeepSeek V4 Pro, and using DistilBERT as the embedding model.
This script outputs the modeled topics.
"""

df = pd.read_csv('00_index.csv') # Change to .csv file containing file names of data
folder_path = 'path_to_data' # Change to folder containing data for topic modeling
docs = []

for index, row in df.iterrows():
    file_name = f"{row['File Name']}"
    file_path = os.path.join(folder_path, file_name)

    if os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            docs.append(f.read())

client = openai.OpenAI(
    api_key='API_KEY', # Change to your active API KEY
    base_url="https://api.deepseek.com")

representation_model = OpenAI(client, model="deepseek-v4-pro", chat=True)

embedding_model = pipeline("feature-extraction", model="distilbert-base-cased")
topic_model = BERTopic(representation_model=representation_model, embedding_model=embedding_model)

topics, probs = topic_model.fit_transform(docs)
```
